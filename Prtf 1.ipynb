## The distribution plot of the total bill

```python
import matplotlib.pyplot as plt
```

```python
sns.distplot(tip['total_bill'],kde=True,bins=100)
plt.show()
```

## The joint plots of the total bill and tip

```python
sns.jointplot(x= 'total_bill',y= 'tip',data= tip,kind='hex')
plt.show()
```

```python
sns.jointplot(x='total_bill',y='tip',data=tip)
plt.show()
```

```python
sns.jointplot(x= 'total_bill',y= 'tip',data= tip,kind='kde')
plt.show()
```

```python
sns.jointplot(x= 'total_bill',y= 'tip',data= tip,kind='reg')
plt.show()
```

## The pairplots of the data 'tip'

```python
sns.pairplot(tip)
plt.show()
```

```python
sns.pairplot(tip,hue='sex')
plt.show()
```

```python
sns.pairplot(tip,hue='sex', palette='seismic')
plt.show()
```

```python
sns.pairplot(tip,hue='sex', palette='viridis')
plt.show()
```

## The Rug Plot of the total bill

## These are some of the categorical plots that can be obtained from the data 'tip'

## Bar Plot: This is the plot of the total bill against sex of clients

```python
sns.barplot(x='sex',y='total_bill',data=tip)
```

## Bar Plot: This is the plot of the total bill against sex of clients using mean as the estimator

```python
sns.barplot(x='sex',y='total_bill',data=tip, estimator=np.mean)
```

## Bar Plot: This is the plot of the total bill against sex of clients using standard deviation as the estimator

```python
sns.barplot(x='sex',y='total_bill',data=tip, estimator=np.std)
```

## Count Plot: This is the plot of the total number of clients according to the sex

```python
sns.countplot(x='sex',data=tip)
```

## Box Plot: This is the plot of the days against the total bill

```python
sns.boxplot(x='day',y='total_bill',data=tip)
```

## Box Plot: This is the plot of the days against the total bill by the sex of the clients 

```python
sns.boxplot(x='day',y='total_bill',data=tip, hue='sex')
```

## Violin Plot: This is the plot of the days against the total bill by the sex of the clients

```python
sns.violinplot(x='day',y='total_bill',data=tip,hue='sex')
```

## Strip Plot: This is the plot of the days against the total bill

```python
sns.stripplot(x='day',y="total_bill",data=tip,jitter=True,dodge=False)
```

## Swarm Plot: This is the plot of the days against the total bill

```python
sns.violinplot(x='day',y='total_bill',data=tip)
sns.swarmplot(x='day', y='total_bill',data=tip,color= 'black')
```

## Factor Plot: This is the plot of the days against the total bill

```python
sns.factorplot(x='day',y='total_bill',data=tip)
```

## Cat Plot: This is the plot of the days against the total bill

```python
sns.catplot(x='day',y='total_bill',data=tip)
```

## Matrix Plot

#### To get the matrix plot. We have to obtain the correlation table from the data 'tip'

```python
tip_corr= tip.corr()
tip_corr
```

## Heatmap of the correlation table of data 'tip'

```python
sns.heatmap(tip_corr,annot=True,cmap='coolwarm')
```

```python
sns.heatmap(tip_corr)
```

#### To show the correlation value

```python
sns.heatmap(tip_corr,annot=True)
```
